```python
# "SaleCityName","ConceptName", "Seasons" sutunlarında kırılınla price ın ortalama değerini inceleyip, price değişkenine göre azalan olarak sıralayıp, yeni bir dataframe oluşturalım
```

```python
agg_df = df.groupby(["SaleCityName","ConceptName", "Seasons"]).agg({"Price": "mean"})
agg_df = agg_df.sort_values("Price", ascending=False)
agg_df
```

```python
agg_df.reset_index(inplace=True)
agg_df.head()
```

```python
# "sales_level_based" adlı yeni bir sutun oluşturalım ve bu sutun SaleCityName ConceptName ve Seasons verilerinin birleştirilmesiyle oluşsun.
```

```python
agg_df["sales_level_based"] = np.array(agg_df["SaleCityName"]) + "_" + np.array(agg_df["ConceptName"]) + "_" + np.array(agg_df["Seasons"])
agg_df["sales_level_based"] = agg_df["sales_level_based"].apply(lambda x: x.upper())
df.head()
```

```python
# price değişkeininden yeni bir kategorik sutun oluşturalım
```

```python
agg_df["Segment"] = pd.qcut(agg_df["Price"], 4 ,labels=["standard","good", "very good","excellent" ])
df
```

```python
agg_df.groupby("Segment").agg({"Price": ["mean","max","sum"]})
```

```python
agg_df.sort_values("Price")
```
